***Data filtering***

```python
df[df['age']>18] #filters data on the basis of age
```

```python
#multiple conditions
#AND Condition
df[
    (df['sex'] == 'F')
    & (df['age'] > 18)
    & (df['age'] < 25)
]
```

```python
#OR condition
df[
    (df['sex'] == 'M')
    | (df['age'] > 18)
]
```
